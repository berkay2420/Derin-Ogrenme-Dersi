```python
low_varience_cols = []
for col in df.columns:
  if df[col].var() == 0:
    low_varience_cols.append(col)
print(low_varience_cols)
```

```python
drop_cols = low_varience_cols + ["Bankrupt?"]
y = df["Bankrupt?"]
X = df.drop(drop_cols, axis=1)
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg_model = LogisticRegression(random_state=42, max_iter=300).fit(X_train_scaled, y_train)
```

```python
y_pred_lr = log_reg_model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report
print(f"accuracy: {accuracy_score( y_test, y_pred_lr)}")
print(classification_report(y_test, y_pred_lr))
```
